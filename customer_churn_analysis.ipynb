```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\sadha\Desktop\CC Analysis\Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn']]
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['Churn'] = df[ 'Churn'].map({'Yes': 1, 'No': 0})
```

```python
df.groupby('gender')['Churn'].sum()
```

```python
df.groupby('gender')['Churn'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Churn rate by gender')
plt.show()
```

```python
df.groupby('Churn')['MonthlyCharges'].mean().plot(kind='bar')
plt.title('Average Monthly Charges by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Average Monthly Charges')
plt.show()
```

```python
df.columns
```

```python
df['PhoneService'] = df['PhoneService'].map({'Yes': 1, 'No': 0})
```

```python
sns.scatterplot(x='PhoneService',y='MonthlyCharges', hue='Churn',data= df)
```

```python
df.dtypes
```

```python
sns.lineplot(x='InternetService',y='MonthlyCharges',hue='Churn',data=df)
```

```python
df.groupby('PaymentMethod')['Churn'].sum().plot(kind='bar')
plt.title('Churn count by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Churn Count')
plt.show()
```

```python
df.groupby('Dependents')['Churn'].sum().plot(kind='bar')
plt.title('Churn count by Dependents')
plt.xlabel('Dependents')    
plt.ylabel('Churn Count')
plt.show()
```
